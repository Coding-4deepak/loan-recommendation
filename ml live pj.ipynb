```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
loan=pd.read_csv('loan.csv')
loan
```

```python
app=pd.read_csv('applicant.csv')
app
```

```python
merge=pd.merge(loan,app,on='applicant_id')
```

```python
merge
```

```python
merge.isnull().sum()/merge.shape[0]*100
```

```python
merge=merge.drop(labels='Other_EMI_plans',axis=1)
```

```python
merge
```

```python
merge=merge.drop(labels='Telephone',axis=1)
merge=merge.drop(labels='Balance_in_existing_bank_account_(lower_limit_of_bucket)',axis=1)
merge=merge.drop(labels='Balance_in_existing_bank_account_(upper_limit_of_bucket)',axis=1)
```

```python
merge
```

```python
merge.isnull().sum()/merge.shape[0]*100
```

```python
merge['Purpose']=merge['Purpose'].fillna(merge['Purpose'].mode()[0])
merge['Property']=merge['Property'].fillna(merge['Property'].mode()[0])
merge['Has_been_employed_for_at_least']=merge['Has_been_employed_for_at_least'].fillna(merge['Has_been_employed_for_at_least'].mode()[0])
merge['Has_been_employed_for_at_most']=merge['Has_been_employed_for_at_most'].fillna(merge['Has_been_employed_for_at_most'].mode()[0])
merge['Savings_account_balance']=merge['Savings_account_balance'].fillna(merge['Savings_account_balance'].mode()[0])
merge=round(merge,0)
```
